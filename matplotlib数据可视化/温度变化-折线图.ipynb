<h3>温度变化显示<h3>  
<h5>画出某城市11点到12点一小时内每分钟的温度变化折线图，温度范围为15--18摄氏度<h5>

```python

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 或者使用其他支持中文的字体
# plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']  # 微软雅黑
# plt.rcParams['font.sans-serif'] = ['KaiTi']  # 楷体

plt.figure(figsize=(20,8))

# 生成数据

x= range (60)
# 在【】列表中，生成60个范围在15--18的随机浮点数  random.uniform（）生成随机数及其范围数字范围
# 构造第一个城市的温度数据
y=[random.uniform(15,18) for i in range(60)]
# 构造北京的温度数据
y_bejing=[random.uniform(5,8) for i in range(60)]
x_time=["11点{}分".format(i) for i in x]
# 生成0到60，步长为1的整数序列
y_ticks=range(50)

# 添加标题

# X轴标题
plt.xlabel("时间")
# Y轴标题
plt.ylabel("温度\摄氏度")
# 坐标系标题
plt.title("温度变化")

# 修改刻度值

# 第一个参数：步长为5 第二个参数：指定跟第一个参数对应的中文
plt.xticks(x[::5],x_time[::5])
# 步长为5
plt.yticks(y_ticks[::5])
plt.ylim(0,40)

# 绘图
plt.plot(x,y,label='上海')
plt.plot(x,y_bejing,color='r',linestyle='--',label='北京')

# 增加图例
plt.legend(loc="best")


plt.show()
```

<h4><span style="color: red">当在多个坐标系里进行绘图的时候，刻度，标题必须分别指定  

plt:对于整体处理，修改  
ax：对于每个坐标系进行处理，修改</span><h4>

```python
# 两个城市的温度，分别在两个坐标系中显示
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 或者使用其他支持中文的字体
# plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']  # 微软雅黑
# plt.rcParams['font.sans-serif'] = ['KaiTi']  # 楷体

fig, ax=plt.subplots(nrows=1,ncols=2,figsize=(20,8))

# 生成数据

x= range (60)
# 在【】列表中，生成60个范围在15--18的随机浮点数  random.uniform（）生成随机数及其范围数字范围
# 构造第一个城市的温度数据
y=[random.uniform(15,18) for i in range(60)]
# 构造北京的温度数据
y_bejing=[random.uniform(5,8) for i in range(60)]
x_time=["11点{}分".format(i) for i in x]
# 生成0到60，步长为1的整数序列
y_ticks=range(50)

# 添加标题

# X轴标题
plt.xlabel("时间")
# Y轴标题
plt.ylabel("温度\摄氏度")
# 坐标系标题
plt.title("温度变化")

# 修改刻度值

# 第一个参数：步长为5 第二个参数：指定跟第一个参数对应的中文
# plt.xticks(x[::5],x_time[::5])
ax[1].set_xticks(x[::5],x_time[::5])
ax[0].set_xticks(x[::5],x_time[::5])
# 步长为5
# plt.yticks(y_ticks[0:40:5])
ax[0].set_yticks(y_ticks[::5])
ax[1].set_yticks(y_ticks[::5])
# 绘图
ax[0].plot(x,y,label='上海')
ax[1].plot(x,y_bejing,color='r',linestyle='--',label='北京')

# 增加图例
ax[0].legend(loc="best")
ax[1].legend(loc="best")


plt.show()
```
